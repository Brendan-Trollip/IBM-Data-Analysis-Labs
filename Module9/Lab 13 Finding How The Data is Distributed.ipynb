### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
print("Column names:\n", df.columns.tolist())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Count missing values across all columns
missing_counts = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n", missing_counts.head(10))

# Employment (categorical)
employment_mode = df["Employment"].mode()[0]
df["Employment"] = df["Employment"].fillna(employment_mode)

# JobSat (categorical)
jobsat_mode = df["JobSat"].mode()[0]
df["JobSat"] = df["JobSat"].fillna(jobsat_mode)

# RemoteWork (categorical)
remotework_mode = df["RemoteWork"].mode()[0]
df["RemoteWork"] = df["RemoteWork"].fillna(remotework_mode)

# Example for numerical column
if "ConvertedCompYearly" in df.columns:
    median_comp = df["ConvertedCompYearly"].median()
    df["ConvertedCompYearly"] = df["ConvertedCompYearly"].fillna(median_comp)

```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
# Value counts for Employment, JobSat, and YearsCodePro
print("Employment distribution:\n", df["Employment"].value_counts(dropna=False))
print("\nJobSat distribution:\n", df["JobSat"].value_counts(dropna=False))
print("\nYearsCodePro distribution:\n", df["YearsCodePro"].value_counts(dropna=False))
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
# Pie chart of JobSat
plt.figure(figsize=(8,8))
df["JobSat"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=90, colormap="viridis")
plt.title("Job Satisfaction Distribution")
plt.ylabel("")
plt.show()

# KDE plot (if JobSat is numeric)
if pd.api.types.is_numeric_dtype(df["JobSat"]):
    plt.figure(figsize=(8,6))
    sns.kdeplot(df["JobSat"].dropna(), shade=True, color="skyblue")
    plt.title("Job Satisfaction KDE Plot")
    plt.xlabel("JobSat")
    plt.show()

```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Split multi-choice responses into lists
df["LangWorkedList"] = df["LanguageHaveWorkedWith"].dropna().apply(lambda x: x.split(";"))
df["LangWantList"] = df["LanguageWantToWorkWith"].dropna().apply(lambda x: x.split(";"))

# Explode into separate rows
worked_df = df.explode("LangWorkedList")
want_df = df.explode("LangWantList")

# Frequency counts
worked_counts = worked_df["LangWorkedList"].value_counts().head(10)
want_counts = want_df["LangWantList"].value_counts().head(10)

# Grouped bar chart
lang_compare = pd.DataFrame({"WorkedWith": worked_counts, "WantToWorkWith": want_counts})
lang_compare.plot(kind="bar", figsize=(12,6))
plt.title("Top Programming Languages: Worked With vs Want To Work With")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Grouped bar chart of RemoteWork by Country
remote_ct = pd.crosstab(df["Country"], df["RemoteWork"])
remote_ct.plot(kind="bar", stacked=True, figsize=(12,6), colormap="coolwarm")
plt.title("Remote Work Trends by Country")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

# Heatmap for clearer visualization
sns.heatmap(remote_ct, cmap="YlGnBu")
plt.title("Remote Work Preferences by Region")
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
# Convert YearsCodePro to numeric
df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0.5,
    "More than 50 years": 51
})
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors="coerce")

# If JobSat is categorical, map to numeric scores
job_sat_map = {
    "Very dissatisfied": 1,
    "Slightly dissatisfied": 2,
    "Neutral": 3,
    "Slightly satisfied": 4,
    "Very satisfied": 5
}
df["JobSatScore"] = df["JobSat"].map(job_sat_map)

# Calculate correlation
corr = df[["YearsCodePro", "JobSatScore"]].corr(method="pearson")
print("Correlation between YearsCodePro and JobSatScore:\n", corr)

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="YearsCodePro", y="JobSatScore", alpha=0.6)
plt.title("Correlation Between Experience and Job Satisfaction")
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Job Satisfaction Score")
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Create cross-tabulation
edu_employment_ct = pd.crosstab(df["EdLevel"], df["Employment"])

# Display the table
print(edu_employment_ct)

# Visualize with stacked bar plot
edu_employment_ct.plot(kind="bar", stacked=True, figsize=(12,6), colormap="viridis")
plt.title("Employment Type by Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()
```
