### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
# Bubble Plot: Age vs. Participation Frequency with Job Satisfaction as bubble size
plt.scatter(
    df["Age"],          # x-axis: numeric age
    df["SOPartFreq"],      # y-axis: frequency of participation
    s=df["JobSat"]*20,     # bubble size scaled by job satisfaction
    alpha=0.5,
    c=df["JobSat"],        # color by job satisfaction
    cmap="viridis"
)
plt.title("Age vs. Participation Frequency (Bubble Size = Job Satisfaction)")
plt.xlabel("Age (numeric)")
plt.ylabel("Participation Frequency")
plt.colorbar(label="Job Satisfaction")
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

plt.scatter(
    df["ConvertedCompYearly"],   # x-axis: compensation
    df["JobSat"],                # y-axis: job satisfaction
    s=df["Age"]*10,           # bubble size scaled by age
    alpha=0.5,
    c=df["Age"],              # color by age
    cmap="viridis"
)
plt.title("Compensation vs. Job Satisfaction (Bubble Size = Age)")
plt.xlabel("Yearly Compensation")
plt.ylabel("Job Satisfaction")
plt.colorbar(label="Age (numeric)")
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here

# Step 1: Split multiple languages into lists
df["LanguageList"] = df["LanguageHaveWorkedWith"].dropna().apply(lambda x: x.split(";"))

# Step 2: Explode into separate rows
lang_age = df.explode("LanguageList")

# Step 3: Group by Age and Language, count frequency
lang_counts = lang_age.groupby(["Age","LanguageList"]).size().reset_index(name="Count")

# Step 4: Bubble plot
plt.scatter(
    lang_counts["Age"],
    lang_counts["LanguageList"],
    s=lang_counts["Count"].astype(float)*5,   # bubble size = frequency
    alpha=0.6,
    c=lang_counts["Count"].astype(float),
    cmap="plasma"
)
plt.title("Programming Language Popularity Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Programming Language")
plt.xticks(rotation=45)
plt.colorbar(label="Respondent Count")
plt.show()


```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
# Average job satisfaction per database + respondent counts
db_stats = df.groupby("DatabaseWantToWorkWith")["JobSat"].mean()
db_counts = df["DatabaseWantToWorkWith"].value_counts()

plt.scatter(
    db_stats.index,
    db_stats.values,
    s=db_counts.loc[db_stats.index].astype(float)*10,   # bubble size = respondent count
    alpha=0.6,
    c=db_counts.loc[db_stats.index].astype(float),
    cmap="coolwarm"
)
plt.title("Preferred Databases vs. Job Satisfaction")
plt.xlabel("Database")
plt.ylabel("Average Job Satisfaction")
plt.xticks(rotation=45)
plt.colorbar(label="Respondent Count")
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
# Average compensation per role + job satisfaction as bubble size
role_comp = df.groupby("DevType")["ConvertedCompYearly"].mean()
role_sat  = df.groupby("DevType")["JobSat"].mean()

plt.scatter(
    role_comp.index,
    role_comp.values,
    s=role_sat.astype(float)*20,   # bubble size = job satisfaction
    alpha=0.6,
    c=role_sat.astype(float),
    cmap="viridis"
)
plt.title("Compensation Across Developer Roles (Bubble Size = Job Satisfaction)")
plt.xlabel("Developer Role")
plt.ylabel("Average Compensation")
plt.xticks(rotation=45)
plt.colorbar(label="Job Satisfaction")
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
# Group by tool and age, count frequency
tools_age = df.groupby(["NEWCollabToolsHaveWorkedWith","Age"]).size().reset_index(name="Count")

plt.scatter(
    tools_age["Age"],
    tools_age["NEWCollabToolsHaveWorkedWith"],
    s=tools_age["Count"].astype(float)*10,   # bubble size = frequency
    alpha=0.6,
    c=tools_age["Count"].astype(float),
    cmap="magma"
)
plt.title("Collaboration Tools by Age")
plt.xlabel("Age Group")
plt.ylabel("Collaboration Tool")
plt.xticks(rotation=45)
plt.colorbar(label="Frequency")
plt.show()
```
